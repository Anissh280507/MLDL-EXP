# Task
Compare the performance of a Decision Tree classifier and a Random Forest classifier on a suitable dataset, including data preparation, model training, evaluation with key metrics (accuracy, precision, recall, F1-score), visualization of their performance, and a final summary of which model performs better.

## Prepare Data

### Subtask:
Load or generate a suitable dataset for classification, split it into training and testing sets, and preprocess if necessary.


**Reasoning**:
The subtask requires loading a dataset and splitting it into training and testing sets. I will import the necessary functions, load the breast cancer dataset, and then split it as instructed.



```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Data loaded and split successfully.")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

## Train Decision Tree

### Subtask:
Train a Decision Tree classifier on the prepared training data.


**Reasoning**:
To train a Decision Tree classifier, I need to import the `DecisionTreeClassifier` class, instantiate it, and then fit it to the training data (`X_train`, `y_train`).



```python
from sklearn.tree import DecisionTreeClassifier

# Instantiate a Decision Tree classifier
dtc = DecisionTreeClassifier(random_state=42)

# Train the Decision Tree classifier
dtc.fit(X_train, y_train)

print("Decision Tree Classifier trained successfully.")
```

## Train Random Forest

### Subtask:
Train a Random Forest classifier on the prepared training data.


**Reasoning**:
To train a Random Forest classifier as per the subtask instructions, I need to import the `RandomForestClassifier` from `sklearn.ensemble`, instantiate it with a `random_state`, and then fit it using the training data (`X_train`, `y_train`).



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load the breast cancer dataset (re-loading to ensure X_train, y_train are defined)
data = load_breast_cancer()
X = data.data
y = data.target

# Split the dataset into training and testing sets (re-splitting to ensure X_train, y_train are defined)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instantiate a Random Forest classifier
rfc = RandomForestClassifier(random_state=42)

# Train the Random Forest classifier
rfc.fit(X_train, y_train)

print("Random Forest Classifier trained successfully.")
```

## Evaluate Models

### Subtask:
Evaluate the performance of both the Decision Tree and Random Forest classifiers on the test set using relevant metrics such as accuracy, precision, recall, and F1-score.


**Reasoning**:
To evaluate the models, I need to make predictions on the test set for both the Decision Tree and Random Forest classifiers. Then, I will import the necessary metrics from sklearn.metrics and calculate the accuracy, precision, recall, and F1-score for each model, printing them to the console.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier

# Instantiate and train Decision Tree classifier (re-added to ensure it's defined)
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)

# Make predictions with the Decision Tree classifier
y_pred_dtc = dtc.predict(X_test)

# Make predictions with the Random Forest classifier
y_pred_rfc = rfc.predict(X_test)

# Evaluate Decision Tree
accuracy_dtc = accuracy_score(y_test, y_pred_dtc)
precision_dtc = precision_score(y_test, y_pred_dtc)
recall_dtc = recall_score(y_test, y_pred_dtc)
f1_dtc = f1_score(y_test, y_pred_dtc)

# Evaluate Random Forest
accuracy_rfc = accuracy_score(y_test, y_pred_rfc)
precision_rfc = precision_score(y_test, y_pred_rfc)
recall_rfc = recall_score(y_test, y_pred_rfc)
f1_rfc = f1_score(y_test, y_pred_rfc)

# Store metrics in dictionaries
dt_metrics = {
    "Accuracy": accuracy_dtc,
    "Precision": precision_dtc,
    "Recall": recall_dtc,
    "F1-score": f1_dtc
}

rf_metrics = {
    "Accuracy": accuracy_rfc,
    "Precision": precision_rfc,
    "Recall": recall_rfc,
    "F1-score": f1_rfc
}

print("Decision Tree Classifier Metrics:")
for metric, value in dt_metrics.items():
    print(f"  {metric}: {value:.4f}")

print("\nRandom Forest Classifier Metrics:")
for metric, value in rf_metrics.items():
    print(f"  {metric}: {value:.4f}")
```

## Visualize Performance

### Subtask:
Generate a plot (e.g., bar chart) comparing the key performance metrics of the Decision Tree and Random Forest models. Ensure legends are provided.


**Reasoning**:
To visualize the performance comparison between the Decision Tree and Random Forest models, I will create a bar chart as specified in the instructions, using `matplotlib` to plot the metrics and `numpy` for array manipulation.



```python
import matplotlib.pyplot as plt
import numpy as np

# Extract metric names and values
metrics = list(dt_metrics.keys())
dt_values = list(dt_metrics.values())
rf_values = list(rf_metrics.values())

# Set up plot parameters
x = np.arange(len(metrics)) # the label locations
width = 0.35 # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Create bars for Decision Tree
rects1 = ax.bar(x - width/2, dt_values, width, label='Decision Tree')
# Create bars for Random Forest
rects2 = ax.bar(x + width/2, rf_values, width, label='Random Forest')

# Add labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Metrics')
ax.set_ylabel('Score')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

plt.tight_layout()
plt.show()

print("Performance comparison plot generated.")
```
