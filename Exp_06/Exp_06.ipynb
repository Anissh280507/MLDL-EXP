# Task
Apply K-Means and Hierarchical Clustering on a generated sample dataset, visualize the results of both algorithms, and then compare their outcomes.

## Generate Sample Data

### Subtask:
Generate a sample dataset (e.g., using `make_blobs` from scikit-learn) suitable for demonstrating clustering algorithms.


**Reasoning**:
First, I'll import the necessary `make_blobs` function from `sklearn.datasets` to generate the sample data for clustering.



```python
from sklearn.datasets import make_blobs

print("make_blobs imported successfully.")
```

**Reasoning**:
Now that `make_blobs` is imported, I will use it to generate the sample dataset with the specified parameters: 300 samples, 2 features, 4 centers, a standard deviation of 0.6, and a random state of 42.



```python
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42, n_features=2)

print(f"Shape of generated features (X): {X.shape}")
print(f"Shape of true labels (y_true): {y_true.shape}")
```

## Apply K-Means Clustering

### Subtask:
Apply the K-Means clustering algorithm to the generated sample dataset. Determine an appropriate number of clusters or try a few different values.


**Reasoning**:
To apply K-Means clustering, I need to import the `KMeans` class, instantiate it with the specified parameters (`n_clusters=4`, `random_state=42`, `init='k-means++'`), fit the model to the features `X`, and then store the predicted cluster labels.



```python
from sklearn.cluster import KMeans

# Instantiate KMeans with 4 clusters and a random state for reproducibility
kmeans = KMeans(n_clusters=4, random_state=42, init='k-means++', n_init=10)

# Fit the KMeans model to the features X
kmeans.fit(X)

# Store the predicted cluster labels
kmeans_labels = kmeans.labels_

print("K-Means model fitted and labels stored.")
print(f"First 10 K-Means labels: {kmeans_labels[:10]}")
```

## Apply Hierarchical Clustering

### Subtask:
Apply Hierarchical Clustering (e.g., Agglomerative Clustering) to the same sample dataset. Experiment with different linkage methods if desired.


**Reasoning**:
To apply Hierarchical Clustering, I need to import the `AgglomerativeClustering` class, instantiate it with specified parameters, fit the model to the features `X`, and then store the predicted cluster labels.



```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

# Regenerate X since it was not defined in the execution context
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42, n_features=2)

# Instantiate AgglomerativeClustering with 4 clusters and 'ward' linkage
hac = AgglomerativeClustering(n_clusters=4, linkage='ward')

# Fit the AgglomerativeClustering model to the features X and predict labels
hac_labels = hac.fit_predict(X)

print("Agglomerative Clustering model fitted and labels stored.")
print(f"First 10 Hierarchical Clustering labels: {hac_labels[:10]}")
```

## Visualize Clustering Results

### Subtask:
Visualize the results of both K-Means and Hierarchical Clustering, displaying the original data points colored by their assigned clusters. Include legends for clarity.


**Reasoning**:
To visualize the clustering results as requested, I need to import the necessary plotting libraries, create a figure with two subplots, and then plot the data points colored by their respective cluster labels for both K-Means and Hierarchical Clustering.



```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Instantiate KMeans with 4 clusters and a random state for reproducibility
kmeans = KMeans(n_clusters=4, random_state=42, init='k-means++', n_init=10)

# Fit the KMeans model to the features X
kmeans.fit(X)

# Store the predicted cluster labels
kmeans_labels = kmeans.labels_

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot K-Means results
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=kmeans_labels, palette='viridis', legend='full', ax=axes[0])
axes[0].set_title('K-Means Clustering')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Feature 2')

# Plot Hierarchical Clustering results
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=hac_labels, palette='viridis', legend='full', ax=axes[1])
axes[1].set_title('Hierarchical Clustering')
axes[1].set_xlabel('Feature 1')
axes[1].set_ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
